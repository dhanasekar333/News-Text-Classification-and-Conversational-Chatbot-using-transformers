```python
from transformers import pipeline
model_name = "textattack/bert-base-uncased-ag-news"
classifier = pipeline("text-classification", model=model_name,top_k=None)
```

```python
text = ["stock market crashed after new policy",
        "the Indian team won the T20 worldcup",
        "NASA is planning a new mission",
        "govt passed law for healthcare reforms"]
```

```python
for t in text:
  predictions = classifier(t)[0]
  print("text: ", t)
  for pred in predictions:
    print(pred['label'],pred['score'])
```

```python
from google.colab import userdata
test_bert = userdata.get('Test_Bert')
```

```python
from huggingface_hub import InferenceClient
model = "mistralai/Mistral-7B-Instruct-v0.3"
```

```python
client = InferenceClient(model=model, token=test_bert)
message = [
  {"role":"system","content":"assume you are an helpful assistant to write emails"},
  {"role":"user","content":input()}
]
response = client.chat_completion(message)
print(response.choices[0].message['content'])
```
